```python
import pandas as pd
```

```python
df = pd.read_csv('Data_cleaning_27102025.csv')

df
```

Quina és la puntuació mitjana donada pels usuaris als allotjaments turístics?

```python
promedio = round(df["review_scores_rating"].mean(), 2)

print(promedio)
```

Quin percentatge d'allotjaments tenen una avaluació general més gran de 80 a cada ciutat?

```python
resultado = (
    df
    .assign(sup_80 = df["review_scores_rating"] > 80)
    .groupby("city", as_index=False)
    .agg(pt_sup_80=("sup_80", lambda x: round(x.mean() * 100, 2)))
    .sort_values("pt_sup_80", ascending=False))

print(resultado)
```
